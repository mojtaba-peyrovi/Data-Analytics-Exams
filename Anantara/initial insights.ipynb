```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
plt.rcParams["patch.force_edgecolor"] = True
```

```python
data = pd.read_excel('allData.xlsx')
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='Status',data=data,palette="rainbow")
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='Status',data=data,palette="rainbow",hue='Country')
```

```python
plt.figure(figsize=(20,8))
sns.boxplot(x='Tier',y='Age',data=data,palette="rainbow")
```

```python
data[["Country", "Status"]].groupby(['Country'], as_index=False).mean().sort_values(by='Status', ascending=False)
```

```python
data[["Tier", "Status"]].groupby(['Tier'], as_index=False).mean().sort_values(by='Status', ascending=False)
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='Status',data=data,palette="rainbow",hue='Tier')
```
