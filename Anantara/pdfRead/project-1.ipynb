```python
import PyPDF2 as pdf
```

```python
pdf_file = open('Resume.pdf','rb')
```

```python
read_pdf = pdf.PdfFileReader(pdf_file,strict=False)

```

```python
jj = read_pdf.getPage(0)
```

```python
print(jj.extractText())
```

```python
xport = pdf.PdfFileWriter()
```

```python
xport.addPage(jj)
```

```python
print(xport)
```
